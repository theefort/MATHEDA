```python
#imports and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

# Current Power

```python
# Load dataset
df = pd.read_csv(r'dataset\current-power.csv',
                    delimiter=",")
df
```

```python
#summary of statistics
df.describe()
```

```python
#covariance of current and power
df.cov()
```

```python
# scatter plot
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.show()
```

```python
# correlation coefficient of current and power
df.corr()
```

```python
#Regresssion line
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#Scatter Plot
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color='#1A1F24',
            alpha=0.6,
            label='Data Points')
#Line Plot
plt.plot(x, line,
        color='#124562',
        label=f'Regression line (r = {r_value:.3f})')

plt.title('Device Current vs Power Respone')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.legend()
plt.show()
```

# NOx Humidity

```python
#Regresssion line
x = ['NOx']
y = ['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#Scatter Plot
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color='#1A1F24',
            alpha=0.6,
            label='DataPoints')
#Line Plot
plt.plot(x, line,
        color='#124562',
        label=f'Regression line (r = {r_value:.3f})')

plt.title('NOx vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NOx (ug/cu.m)')
plt.legend()
plt.show()
```

```python
#Regresssion line
x = ['NO2']
y = ['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#Scatter Plot
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color='#1A1F24',
            alpha=0.6,
            label='DataPoints')
#Line Plot
plt.plot(x, line,
        color='#124562',
        label=f'Regression line (r = {r_value:.3f})')

plt.title('NO2 vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('CO2 (ug/cu.m)')
plt.legend()
plt.show()
```
